```python
class Reducer:
    def __init__(self, init_vec, obj):
        self.init_vec = init_vec
        self.obj = obj
        self.mask = jnp.ones_like(init_vec).astype(bool)
    
    def new_obj(self, vector, threshold, **kwargs):
        #assert self.mask.astype(int).sum() == vector.shape[0]
        mapped = self._map_vector(vector)
        mask = jnp.abs(mapped) > threshold
        self._update_mask(mask)
        def obj(reduced_vector, **kwargs):
            z = self._map_vector(reduced_vector)
            return self.obj(z, **kwargs)
        return obj, mapped[mask,]
        
    def _map_vector(self, reduced_vector):
        z = jnp.zeros_like(self.init_vec)
        return z.at[self.mask,].set(reduced_vector)
    
    def _update_mask(self, mask):
        self.mask = self.mask & mask
        
        
def np_l0(params, lamb=1e-3):
    l0 = lamb*jnp.sum(params**2)

    params = unflatten(spec, params)
    
    print(mse_fn(params, key), l0)#*l0_fn(params, key))
    return float(loss(params, key) + l0)
```
